```python
from pynance.exchange import EuronextBrussels,EuronextParis
from pynance.broker import KeytradeBroker
from pynance.utils.helpers import millify
```

```python
nxt = EuronextBrussels()
nxt.scrape_tickers()
```

```python
# Validate and compute DCF
for t in nxt.tickers[:10]:
    valuation = t.compute_dcf()
    yf_valuation = t.get_yf_marketcap() or 0
    print(t.yf_symbol,millify(valuation),millify(yf_valuation))
```
